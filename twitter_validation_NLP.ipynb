```python
df['feedback']=df['feedback'].map({'Positive':1,'Negative':-1,'Neutral':0})
df
```

```python
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
tweets=df.text
tweets
```

```python
#tokenization
from nltk.tokenize import word_tokenize
from nltk.tokenize import TweetTokenizer #it is another tokenizer
tk=TweetTokenizer()
tweets=tweets.apply(lambda x:tk.tokenize(x)) #apply is a keyword used before lambda
tweets=tweets.apply(lambda x:' '.join(x)) #this is done to easily handle things
tweets
```

```python
import re
tweets=tweets.str.replace('[^a-zA-Z0-9]',' ',regex=True)
tweets
```

```python
from nltk.tokenize import word_tokenize
from nltk.tokenize import TweetTokenizer
tweets=tweets.apply(lambda x:' '.join([w for w in tk.tokenize(x) if len(w)>=3])) #to collect words greater than 3
tweets
```

```python
#stemming
from nltk.stem import SnowballStemmer
stemmer=SnowballStemmer('english')
tweets=tweets.apply(lambda x:[stemmer.stem(i.lower()) for i in tk.tokenize(x)]).apply(lambda x:' '.join(x))
tweets
```

```python
from nltk.corpus import stopwords
sw=stopwords.words('english')
sw
# lst=[tweets for tweet in tweets if tweet not in sw]
# lst
tweets=tweets.apply(lambda x:[i for i in tk.tokenize(x) if i not in sw]).apply(lambda x:' '.join(x))
tweets
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vec=TfidfVectorizer()
data=vec.fit_transform(tweets)
data
```

```python
print(data)
```

```python
y=df['feedback'].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(data,y,test_size=0.3,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
knn=KNeighborsClassifier(n_neighbors=3)
nb=MultinomialNB()
svm_model=SVC()
dtc=DecisionTreeClassifier(criterion='entropy')
rfc=RandomForestClassifier(n_estimators=100,random_state=42)
models=[knn,nb,svm_model,dtc,rfc]
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
for model in models:
  print("Model started.............")
  model.fit(x_train,y_train)
  print("Predicting values.............")
  y_pred=model.predict(x_test)
  print("Confusion matrix............")
  cm=confusion_matrix(y_test,y_pred)
  print(cm)
  print("Accuracy score................")
  score=accuracy_score(y_test,y_pred)
  print(score)
  print()
```
