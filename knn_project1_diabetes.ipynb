```python
x=df.iloc[:,:-1].values #to print x as an array
x #x(input) should always be 2D
#printing x as array is compulsory
```

```python
y=df.iloc[:,-1].values
y  #y(output) should always be 1D
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1) #test_data is 30% of total
x_train
```

```python
#normalization step
#in this dataset, glucose has high value and diabetespedigreefunction has
#lower value. As a result, the machine will predict diabetes positive for
#high value of diabetes and diabetes negative for low value of diabetes.
#so we have to normalize the data to the same range.

#two types of normalization:
#standard scalar and min_max scalar
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train) #by fitting we mean training the model with the values and transform means
                                      #computing using the equation and scaling the values
x_test=scaler.transform(x_test)
```

```python
import math
k_value=math.sqrt(len(y_test))
k_value
```

```python
#fitting/training the model
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=15)
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)
y_pred
```

```python

df_x_test = pd.DataFrame(x_test)
df_knn=pd.DataFrame({
  "Glucose": df_x_test.iloc[:, 0],
  "Outcome": y_pred
})
df_knn
```

```python
#for classification model, to check the performance of the model
#there are mainly 4 methods:
#1)Accuracy Score - useful when the data is balanced
#2)Recall
#3)Precision
#4)F1_score

#these performance methods work on confusion matrix
#4 technical terms:
#TP-True Positive;y_test=y_pred=1
#TN-True Negative;y_test=y_pred=0
#FP-False Positive;y_test=0,t_pred=1
#FN-False Negative;y_test=1,y_pred=0

#Accuracy Score eqn:
#TP+TN/TP+TN+FP+FN

#CONFUSION MATRIX :
#[TP FP]
#[FN TN]

#recall,precision,f1_score are all suitable when the data is imbalanced
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
#accuracy = accuracy_score(y_test, y_pred)
# print("Accuracy:", accuracy)
```
