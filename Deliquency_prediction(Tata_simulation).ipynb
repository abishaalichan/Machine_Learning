FILLING MISSING VALUES BY IMPUTATION METHOD USING MEAN

```python
df['Income'] = df['Income'].fillna(df['Income'].mean())
df['Credit_Score'] = df['Credit_Score'].fillna(df['Credit_Score'].mean())
df['Loan_Balance'] = df['Loan_Balance'].fillna(df['Loan_Balance'].mean())
```

```python
df.isna().sum()
```

REPLACING WRONG FORMAT DATA

```python
df['Employment_Status'] = df['Employment_Status'].replace('employed', 'Employed')
```

```python
df['Employment_Status'].unique()
```

```python
df['Delinquent_Account'].value_counts()
```

EXTRACTING KEY FEATURES NECESSARY FOR MODELING

```python
df1 = df[['Age','Income','Credit_Score','Credit_Utilization','Missed_Payments','Debt_to_Income_Ratio','Delinquent_Account']]
```

SPLITTING INTO INPUT AND OUTPUT FEATURES

```python
X=df1.drop('Delinquent_Account',axis=1)
y=df1['Delinquent_Account']
```

PERFORMING OVERSAMPLING USING SMOTE TECHNIQUE TO MAKE THE DATASET BALANCED

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
y_resampled.value_counts()
```

SPLITTING INTO TRAINING AND TESTING DATA

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X_resampled,y_resampled,test_size=0.2,random_state=42)
```

MODEL CREATION

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(random_state=42)
model.fit(x_train, y_train)
```

PERFORMANCE EVALUATION

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

y_pred = model.predict(x_test)

print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
print(f"Precision: {precision_score(y_test, y_pred)}")
print(f"Recall: {recall_score(y_test, y_pred)}")
print(f"F1 Score: {f1_score(y_test, y_pred)}")
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

## Data Analysis Key Findings
